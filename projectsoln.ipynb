```python
import pandas as pd 
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
import seaborn as sns

```

```python
df=pd.read_csv("innercity.csv")

r=df.shape[0]
c=df.shape[1]
print(r,c)
df.head(10)
```

```python
df=df.drop(columns=['cid','living_measure15','lot_measure15'],axis=1)
print(df.shape[0],df.shape[1])
df.head(10)
```

```python
# sns.pairplot(df)
```

```python
df['dayhours'] = pd.to_datetime(df['dayhours'])

df['year'] = df['dayhours'].dt.year
df['month'] = df['dayhours'].dt.month
df['day'] = df['dayhours'].dt.day
df['day_of_week'] = df['dayhours'].dt.dayofweek

df=df=df.drop(columns='dayhours',axis=1)
df.info()
df.head(10)
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as sfs 
X=df.drop('price',axis=1)
y=df['price']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=1)
linR=LinearRegression()
sfs1=sfs(linR,k_features=14,forward=True,scoring="r2",cv=14)
sfs1.fit(X_train,y_train)

columnlist=list(X_train.columns)
feature_cols=list(sfs1.k_feature_idx_)
subset=[columnlist[i] for i in feature_cols]
print(subset)
```

```python
#['room_bed', 'room_bath', 'living_measure', 'coast', 'sight', 'condition', 'quality', 'basement', 'yr_built', 'zipcode', 'lat', 'long', 'furnished', 'year']


mpg_df=df[subset].copy()
mpg_df['basement']=mpg_df['basement'].replace(0, np.median(mpg_df.loc[mpg_df['basement'] != 0, 'basement']))
mpg_df = pd.concat([mpg_df, df['price']], axis=1)
mpg_df
```

```python
from sklearn.model_selection import RandomizedSearchCV

X=mpg_df.drop('price',axis=1)
y=mpg_df['price']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=1)
linR=LinearRegression()
param_dist = {
    'fit_intercept': [True, False],
    'positive': [True, False],
    'copy_X': [True], 
}

randomcv=RandomizedSearchCV(linR,param_distributions=param_dist,n_iter=20)
randomcv.fit(X_train,y_train)
best_params=randomcv.best_params_

print("Possible parameters:",linR.get_params())
print("Best ones from randomcv:",best_params)

randomcv.cv_results_['mean_test_score']

```

```python
#linear regression scores with randomsearchcv:
best_linR = LinearRegression(**best_params)

# Train the model
best_linR.fit(X_train, y_train)

# Predict on training and testing sets
y_train_pred = best_linR.predict(X_train)
y_test_pred = best_linR.predict(X_test)

# Calculate scores
train_score = best_linR.score(X_train, y_train)  
test_score = best_linR.score(X_test, y_test)

print([train_score,test_score])
```

```python
#Polynomial degree:
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

poly=PolynomialFeatures(degree=2,include_bias=False)

x_polytrain=poly.fit_transform(X_train)

pipeline = Pipeline([
    ('poly', PolynomialFeatures(include_bias=False)),  
    ('linreg', LinearRegression())                    
])

param_dist = {
    'poly__degree': [2, 3],                      
    'linreg__fit_intercept': [True, False],         
    'linreg__positive': [False],             
}

randomcv = RandomizedSearchCV(
    pipeline,
    param_distributions=param_dist,
    n_iter=6,
    cv=4, 
    random_state=42
)

randomcv.fit(X_train, y_train)

dtreebest=randomcv.best_params_
print("Best parameters:", dtreebest)
print("Best score (Negative MSE):", randomcv.best_score_)

score = randomcv.score(X_test,y_test) 
print("Model RÂ² score:", score)

test_score_percentage = score * 100

print("Test RÂ² score accuracy (out of 100%):", test_score_percentage)
```
